### jsonファイルの編集

```python
# テキストを形態素分析してくれる
import MeCab
# 自然言語処理用ライブラリ・単語のベクトル表現を獲得する
import gensim
from collections import Counter
import re
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./meta.json", delimiter = "\n", header=None)
```

```python
tag = []
for i in df.index:
    meta = json.loads(df.iloc[i].values[0])
    meta['tags']
    tag.append(meta['tags'])
```

```python
str_tag = str(tag)
```

```python
tag_change = str_tag.replace(',','').replace('[','').replace(']','').replace("'","").replace(' ','\n')
```

```python
mecab = MeCab.Tagger("-Owakati -d /usr/local/lib/mecab/dic/mecab-ipadic-neologd/")
t = mecab.parse(tag_change)
```

```python
t1 = t.lower().replace(' / ','/').replace('-','')
```

```python
words = re.split(r'\s|\,|\.|\(|\)', t1.lower())
count = Counter(words)
```

```python
print(count)
```

```python
for word, cnt in count.most_common():
    print(word) 
    print(cnt)
```
